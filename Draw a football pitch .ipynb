Para poder importar la librería *``mplsoccer``* tendrás que haberla instalado inicialmente con *``pip install mplsoccer``* en tu ordenador. Una vez instalada podrás importarla.

```python
from mplsoccer.pitch import Pitch
```

## Dibujamos el campo de fútbol

Con el método Pitch pasándole por parámetro las diferentes característica lo guardamos en una variables. Para poder ver el campo creado aplicamos el método draw() a la variables donde hemos guardado la función Pitch.

```python
pitch = Pitch(figsize=(12,8), pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```

### Cambiamos la estética del campo
Si dejamos el parametro *``stripe=False``* pasaremos a tener el todo el campo de color verde, en vez de la franjas de colores claro y oscuro en el campo anterior.

```python
pitch = Pitch(figsize=(12,8),pitch_color='grass', line_color='white', stripe=False)
fig, ax = pitch.draw()
```

### Cambiamos el color 
Con *``pitch_color``* podemos poner el color que queramos, como por ejemplo el blanco.

```python
pitch = Pitch(figsize=(12,8),pitch_color='white', line_color='black', stripe=False)
fig, ax = pitch.draw()
```

Con los atributos length y width podemos saber el largo y ancho del campo 

```python
pitch.length
```

```python
pitch.width
```

### Graficamos la mitad del campo
Si queremos tener solo la mitad del campo con el parámetro *``view='half'``* y con *``orientation='vertical'``* podemos conseguirlo

```python
pitch = Pitch(figsize=(12,8),pitch_color='grass',view='half', line_color='white', orientation='vertical', stripe=False)
fig, ax = pitch.draw()
```
